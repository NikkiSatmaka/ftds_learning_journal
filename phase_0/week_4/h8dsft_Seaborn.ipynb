## Importing Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

sns.set_theme(style='darkgrid', palette='deep')

pd.set_option('display.precision', 2)

%matplotlib inline
```

## Data Loading

```python
df = pd.read_csv('data/FIFA_data.csv', index_col=[0])

df.head()
```

## Data Visualization

### Average Overall Point by Position

```python
plt.figure(figsize=(15, 5))
sns.barplot(data=df, x='Position', y='Overall')
plt.title("Average Overall Point by Position")
plt.show()
```

### Distribution of players' potential

```python
plt.figure(figsize=(15, 5))
sns.histplot(data=df, x='Potential', bins=50)
plt.title("Distribution of players' potential")
plt.show()
```

### Distribution of players' stamina

```python
plt.figure(figsize=(15, 5))
sns.kdeplot(data=df, x='Stamina')
plt.title("Distribution of players' Stamina")
plt.show()
```

### Distribution of players' finishing

```python
sns.displot(data=df, x='Finishing', bins=50, kde=True)
plt.title("Distribution of players' Finishing")
plt.show()
```

### Composition of players' position

```python
plt.figure(figsize=(15, 15))
plt.pie(x=df.groupby('Position')['ID'].count(), labels=df.groupby('Position')['ID'].count().index)
plt.title("Composition of players' position")
plt.show()
```

### Pairplot between 'Age', 'Overall', 'Potential', 'Special', 'International Reputation'

```python
sns.pairplot(data=df.select_dtypes(include=np.number).iloc[:100, 1:6])
plt.show()
```

### Distribution of players' Stamina across Positions

```python
plt.figure(figsize=(15, 5))
sns.violinplot(data=df, x='Position', y='Stamina')
plt.title("Distribution of players' Stamina across Positions")
plt.show()
```
