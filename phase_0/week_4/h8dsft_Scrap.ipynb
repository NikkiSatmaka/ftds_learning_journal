Extract Title, Rating, Description, Votes, and Poster

https://www.imdb.com/search/title/?genres=animation

```python
import requests
import numpy as np
import pandas as pd

from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
url = 'https://www.imdb.com/search/title/?genres=animation'
res = requests.get(url)
res.raise_for_status()
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
cols = ['Title', 'Rating', 'Desc', 'Votes', 'Poster']
df = pd.DataFrame(columns=cols)

for div_tag in soup.find_all("div", {"class":"lister-item mode-advanced"}):
    title = div_tag.find("h3", {"class":"lister-item-header"}).find(href=True)
    rating = div_tag.find("div", {"class":"inline-block ratings-imdb-rating"})
    desc = div_tag.find_all("p", {"class":"text-muted"})[1]
    votes = div_tag.find("span", {"name":"nv"})
    poster = div_tag.find("img", {"class": "loadlate"})

    # Check for any None values, otherwise, get the data
    if title is not None:
        title = title.get_text().strip()

    if rating is not None:
        rating = rating.get("data-value")

    if desc is not None:
        desc = desc.get_text().strip()

    if votes is not None:
        votes = votes.get("data-value")

    if poster is not None:
        poster = poster.get('loadlate')

    new_entry = [title, rating, desc, votes, poster]
    new_entry = pd.DataFrame(dict(zip(cols, new_entry)), index=[0])

    df = pd.concat([df, new_entry])

df = df.reset_index(drop=True)
```
